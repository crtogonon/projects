# Results / Predictions

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import BaggingClassifier

pd.set_option('display.max_columns', None)
```

## Setting up X and y for training set

```python
nba=pd.read_csv('/Users/ct/DSI-NYC-4/projects/capstone/part-03/nonulls_2.csv')
nba.head()
```

```python
X_train=nba.iloc[:,2:]
X_train.head()
```

```python
X_scaler = StandardScaler()
Xt_scaled = X_scaler.fit_transform(X_train)
```

```python
def grader(x):
    if x>17:
        return 'good'
    elif x<=17 and x>=12:
        return 'average'
    elif x<12:
        return 'bad'
```

```python
nba['grade']=nba['PER'].apply(lambda x: grader(x))
nba.head()
```

```python
nba['grade'].value_counts()
```

```python
y=nba['grade']
y.head()
```

# Setting up and Fitting Model

```python
gb=GradientBoostingClassifier(max_features='log2', min_samples_leaf=2, min_samples_split=4,n_estimators=200)
b1 = BaggingClassifier(gb)
b1.fit(Xt_scaled,y)
```

# Setting up X for prediction set

```python
cp=pd.read_csv('/Users/ct/DSI-NYC-4/projects/capstone/part-03/cp_formatted.csv')
cp.head()
```

```python
X=cp.iloc[:,1:]
X.head()
```

```python
X_scaled = X_scaler.fit_transform(X)
```

# Predictions

```python
pred=b1.predict(X_scaled)
```

```python
len(pred)
```

```python
pd.to_pickle(b1, 'model_v3.pkl')
```

```python
cp['Pred']=pred
```

```python
cp
```

```python
# checking prediction distribution
cp['Pred'].value_counts()
```

```python
cp.to_csv('cp_pred_2.csv',index=False)
```

```python
# getting predicted probabilities
probs=pd.DataFrame(b1.predict_proba(X_scaled))
probs.head()
```

```python
# renaming columns for predicted probabilities
probs.columns = ['avg', 'bad','good']
```

```python
probs.head()
```

```python
# joining probabilities to the dataframe
cp=pd.concat([cp,probs], axis=1)
```

```python
cp.to_csv('cp_pred_3.csv',index=False)
```
