# Project: Billboard Top 100 from the year 2000

### Problem statement:
For this project, I am analyzing a data set that contains information for all of the songs that made the Billboard Top 100 chart in the year 2000. The dataset shows the year, artist name, track name, length of track, song genre, date the song entered the chart, date the song peaked, as well as the chart position for each week the song was in the top 100. The goal for the project will be to analyze the information found in the dataset and determine what makes a song popular.

### Risks and Assumptions:
After reading in the data set, the data first needs to be cleaned. There could be issues with the data an each column that I would need to investigate. Things such as possible duplicate artist names due to misspellings or alternate spellings, fixing the format of the time column, and ensuring that the data in each column is of the appropriate data type necessary for me to perform analysis on.

For this data set I will be making the assumption that all of the songs are properly labeled with their appropriate genre. It is unclear what genre criteria Billboard used for this chart, and any reclassification of the tracks on my part would only be inserting my own bias into the data set.

### Data Cleaning:

```python
#Replacing the asterisks in the week columns with null values

def replace_nulls(value):
    if value == '*':
        return np.nan
    else:
        return value
    
bb=bb.applymap(replace_nulls)
bb
```

```python
# checking the datatypes after replacing the asterisks
bb.dtypes
```

```python
# Checking the value counts to see if there are any duplicate genres listed.
# It looks like R&B is listed twice in different ways.

bb['genre'].value_counts()
```

```python
# Fixing the column names  for the R&B genre so that they are all classified with the same tag.
# I went with "R&B"

def fixrnb(x):
    if x=="R & B":
        x="R&B"
    else:
        pass
    return x
        
bb['genre']=bb['genre'].apply(fixrnb)

bb['genre'].value_counts()
```

```python
# Inspecting the entire list of artist names to make sure that no artists are listed twice.
# After inspection, the list looks good.

set(bb['artist.inverted'])
```

```python
# Split the time column in order to create new columns that would help me evaluate the data better.
# I created columns for the number of minutes of each track, the seconds, the minute values converted into seconds,
# and the total amount of seconds for each song.

timesplit=[]
time2=[]

# Splitting the time column into a list of lists in 'timesplit'
for x in bb['time']:
    timesplit.append(x.split(','))

# I only need the first two values of each list produced so I put those into a new list
for y in timesplit:
    time2.append([int(y[0]),int(y[1])])

minutes=[]
seconds=[]
minutestoseconds=[]
totalseconds=[]

for z in time2:                    # Making a list of all of the minute values
    minutes.append(z[0])
    
for z in time2:                    # Making a list of all of the second values
    seconds.append(z[1])

for z in minutes:                  # Making a list of all of the minutes converted into seconds
    minutestoseconds.append(z*60)

# Creating the new columns related to the song length:
bb['minutes']=pd.Series(minutes)
bb['minutes to seconds']=pd.Series(minutestoseconds)
bb['seconds']=pd.Series(seconds)
bb['total seconds']=bb['minutes to seconds']+bb['seconds']
```

```python
# Checking the output of one cell in the date entered column to try to determine how best to
# convert the entries into data I can work with

bb['date.entered'][2]
```

```python
# Splitting the date.entered column into month, day, and year and creating new columns for each

desplit=[]

for x in bb['date.entered']:    # splitting the date entered column and putting it into desplit
    desplit.append(x.split())
    
for x in desplit:               # removing the ',' from the day number values
    x[1]=x[1][:-1]
    
for x in desplit:               # converting the values for day number and year into integers
    x[1]=int(x[1])
    x[2]=int(x[2])
    
demonth=[]
for x in desplit:               # creating a list of all of the month values
    demonth.append(x[0])
    
deday=[]
for x in desplit:               # creating a list of all of the day values
    deday.append(x[1])
    
deyear=[] 
for x in desplit:               # creating a list of all of the year values
    deyear.append(x[2])
    
# Creating all of the new columns for the entered date
bb['entered.month']=pd.Series(demonth)
bb['entered.day']=pd.Series(deday)
bb['entered.year']=pd.Series(deyear)
```

```python
# The same process used above was used for the dates in the "date peaked" column

dpsplit=[]

for x in bb['date.peaked']:
    dpsplit.append(x.split())
    
for x in dpsplit:
    x[1]=x[1][:-1]
    
for x in dpsplit:
    x[1]=int(x[1])
    x[2]=int(x[2])
    
dpmonth=[]
for x in dpsplit:
    dpmonth.append(x[0])
    
dpday=[]
for x in dpsplit:
    dpday.append(x[1])
    
dpyear=[]
for x in dpsplit:
    dpyear.append(x[2])
    
bb['peaked.month']=pd.Series(dpmonth)
bb['peaked.day']=pd.Series(dpday)
bb['peaked.year']=pd.Series(dpyear)
```

```python
# Creating a column that shows how many weeks each song was on the Top 100 chart. Used the count function.

bb['weeks.on.chart']=bb.loc[:,'x1st.week':'x76th.week'].count(axis=1)
```

```python
# Converting all of the data in the week columns into floats so I can run functions on them:
bb.loc[:,'x1st.week':'x76th.week']=bb.loc[:,'x1st.week':'x76th.week'].astype(float)

# Creating columns that show the mean, min, and max position on the chart for each song:
bb['average.position']=bb.loc[:,'x1st.week':'x76th.week'].mean(axis=1)
bb['highest.position']=bb.loc[:,'x1st.week':'x76th.week'].min(axis=1)
bb['lowest.position']=bb.loc[:,'x1st.week':'x76th.week'].max(axis=1)
```

```python
# Creating a dataframe using value_counts to show how many weeks a song was at each position on the chart
bb2=bb.loc[:,'x1st.week':'x76th.week']
bb3=bb2.apply(pd.Series.value_counts, axis=1)

# Creating a column that shows how many weeks a song was number 1 on the chart
bb['weeks.at.1']=bb3.iloc[:,0]

# Showing the first few rows of the dataframe created to show the position counts
bb3.head()
```

```python
# Creating a column that shows the number of weeks in the Top 5 for each song:
bb['weeks.top.5']=bb3.iloc[:,0:5].sum(axis=1)
```

```python
# Creating a column that shows the number of weeks in the Top 10 for each song:
bb['weeks.top.10']=bb3.iloc[:,0:10].sum(axis=1)
```

```python
# Creating a column that shows the number of weeks in the Top 25 for each song:
bb['weeks.top.25']=bb3.iloc[:,0:25].sum(axis=1)
```

### Hypothesis:

I've decided to test the hypothesis that the average entry position for songs that get to number 1 is the same as the average entry position for all songs that make the top 100.

The null hypothesis would be that there is no difference between the average 1st week position of songs that made it to number 1 and the average 1st week position of all songs in the top 100.

```python
# Calculated the average starting position for all songs that made the top 100.

bb['x1st.week'].mean()
```

```python
from scipy import stats

# Running the t test for all of the songs that made it to number 1
# Rounded the mean result from the previous cell up to 80
print stats.ttest_1samp(bb1['x1st.week'],80.0)
```

For a significance value of 0.05, the resulting pvalue for our data is well below the threshold which means that we reject the null hypothesis. The average entry position for songs that reach number 1 is likely not the same as the average entry position for all songs that reach the top 100.
